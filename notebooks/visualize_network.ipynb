```python
import matplotlib.pyplot as plt
import networkx as nx
import random
```

```python
# Import the network gexf file
G = nx.read_gexf('../data/processed/network/network.gexf')
```

```python
# Remove nodes with no edges
G.remove_nodes_from(list(nx.isolates(G)))

# Sample a smaller subgraph to visualize, for example 1000 nodes
# sampled_nodes = random.sample(G.nodes(), 3000)
# subgraph = G.subgraph(sampled_nodes)
subgraph = G

# Generate a layout for the subgraph
pos = nx.kamada_kawai_layout(subgraph)
# pos = nx.spring_layout(subgraph, k=0.1, iterations=20)

# Draw the nodes and the edges (tune the size and opacity to make the graph readable)
plt.figure(figsize=(10, 10))
# nx.draw_networkx_nodes(subgraph, pos, node_size=20, alpha=0.6)
# nx.draw_networkx_edges(subgraph, pos, alpha=0.2)
nx.draw(subgraph, pos, edge_color='gray', style='solid', with_labels=False, node_size=5, font_size=8, width=0.2)

# Hide the axis
plt.axis('off')

# Show the plot
plt.show()
```
